### Function to get average number of words per sentence from corpus

```python
# assuming we are only using inaugural
def get_average_num_of_word_per_sentence(filename):
    len_of_words = len(inaugural.words(filename))
    len_of_sents = len(inaugural.sents(filename))
    return len_of_words / len_of_sents
```

### Get average num of words / sentence in all inaugural speeches

```python
total_words = 0
total_sentences = 0
total_avg_word_per_sentence = 0
for each_file in inaugural.fileids():
    total_words += len(inaugural.words(each_file))
    total_sentences += len(inaugural.sents(each_file))
    avg_word_per_sentence = get_average_num_of_word_per_sentence(each_file)
    total_avg_word_per_sentence += avg_word_per_sentence
    print('Filename:', each_file, '\tAvg Word/Sentence:', avg_word_per_sentence)
```

## Part 4 — Bigrams

```python
text = "I think it might rain today." 
```

```python
tokens = nltk.word_tokenize(text)
tokens
```

**Bigrams are pairs of adjacent word from our original phrase**

```python
bigrams = nltk.bigrams(tokens)
for pair in bigrams:
    print(pair)
```

**Trigrams are triplets of adjacent word from our original phrase**

```python
trigrams = nltk.trigrams(tokens)
for triplets in trigrams:
    print(triplets)
```

**ngrams allow you to split the phrase into any numbers of adjacent word from our original phrase**

```python
from nltk.util import ngrams
```

**Example: Quadrauple Grams**

```python
text2 = "If it is nice out, I will go to the beach."
tokens = nltk.word_tokenize(text2)
# you can pass in any number you want
quadrapulegrams = ngrams(tokens, 4)
for my_gram in quadrapulegrams:
    print(my_gram)
```

**An ngrams function to make your life easier**

```python
def n_grams(text, n):
    tokens = nltk.word_tokenize(text)
    grams = ngrams(tokens, n)
    return grams

# please note to not confuse ngrams with n_grams here, the result wld be diff
grams = n_grams("This is a long string and I want to split it into bigrams", 2)
for gram in grams:
    print(gram)
```

## Part 2 — Normalizing
- Text normalization is the process of transforming a text into a canonical (standard) form. For example, the word “gooood” and “gud” can be transformed to “good”, its canonical form. Another example is mapping of near identical words such as “stopwords”, “stop-words” and “stop words” to just “stopwords”

```python
md = nltk.corpus.gutenberg.words("melville-moby_dick.txt")
md_22 = md[:22] # get first 22 characters
md_22
```

**Normalize by removing alphabetical character**

```python
# here, we can filter away those non alphabetical special characters etc. using the .isalpha() method
for word in md_22:
    # The isalpha() method returns True if all the characters are alphabet letters (a-z).
    if word.isalpha(): print(word)
```

**Extending the normalization to make everything that is alpha character lowercase**

```python
# we can also convert everything to lower or uppercase using .lower() and .upper() method
# combining the both allow us to get a more normalized list of words
# if a word is alpha, convert it to lower
md_norm = [word.lower() for word in md_22 if word.isalpha()]
md_norm
```

**Stemming — normalize the affixes, i.e. convert plural forms of words into singular etc. — may not work all the time depending on the algo**
- An affix is a word part added to the beginning or the end of the word. An affix can change the meaning of the word, and also can be used to change the part of speech of the word.
- There are two types of affixes: prefixes and suffixes. A prefix is added to the beginning of the word. A suffix is added to the end of the word.

<u>Common Prefixes</u>
- re- (again)
- un- (not)
- dis- (not)

Example — (dis)like — i.e. not like

<u>Common Suffixes</u>
- able (can be done, able to)
- ful (full of)
- ing (verb ending, progressive tense)

Example: care(less) — i.e. without care

Source: https://www.softschools.com/examples/literary_terms/affix_examples/698/

```python
porter = nltk.PorterStemmer()
```

```python
my_list = ["cat", "cats", "lie", "lying", "run", "running", "city", "cities", "month", "monthly", "woman","women"]
```

```python
# most seems to work, but some things like city, monthly, and woman doesn't work
for word in my_list: print(porter.stem(word))
```

**Lemmetization — can be used to remove the affixes too — but computationally expensive**
- Lemmatization is the process of converting a word to its base form. The difference between stemming and lemmatization is, lemmatization considers the context and converts the word to its meaningful base form, whereas stemming just removes the last few characters, often leading to incorrect meanings and spelling errors.

```python
wmlem = nltk.WordNetLemmatizer()
for word in my_list: print(wmlem.lemmatize(word))
```

**Basic Example — A simple sentiment analysis function you can use**

```python
def get_sentiment_from_text(text):
    # split text into sentences
    text_sent = nltk.sent_tokenize(text)
    print(len(text_sent), "sentence(s) detected...")
    # for each sentence
    for sentence in text_sent:
        print()
        print("Analayzing Sentence: " + sentence + " ...")
        n_count = 0 # num of negative words
        p_count = 0 # num of positive words
        # split each sentence into words
        sent_words = nltk.word_tokenize(sentence)
        for word in sent_words: # loop through each words
            if word in positive_words: 
                print("- Identified '" + word + "' as positive")
                p_count += 1 # if word is inside positive word list, add p_count
            if word in negative_words: 
                print("- Identified '" +  word + "' as negative")
                n_count += 1 # if word is inside negative word list, add p_count
    
        # the tutorial's algo — logic is a little flawed
        print("=== Tutorial Algo ===")
        # if only got positive and no negative words = positive
        if(p_count > 0 and n_count == 0):
            print(sentence, '|', "Positive")
        # if there's odd number of negative words = negative
        # doesn't make sense to me lol, if u got 3 neg word 15 positive words, it'll still be negative tf.
        elif(n_count % 2 > 0):
            print(sentence, '|', "Negative")
        elif(n_count % 2 == 0 and n_count > 0):
            print(sentence, '|', "Positive")
        else:
            print(sentence, '|', "Neutral")
                
        print()
        # my custom algo
        print("=== My Custom Algo ===")
        if p_count == n_count: 
            print(sentence, '|', "Neutral")
            # return "Neutral"
        elif p_count > n_count : 
            print(sentence, '|', "Positive")
            # return "Positive"
        else: 
            print(sentence, '|', "Negative")
            # return "Negative"
        
```

**A real life example**

```python
# modify the previous function for this purpose — I'll focus on my custom algo rather than the tutorial algo (cause i think mine makes more sense)
def get_sentiment_from_text(text):
    sentiment_score_list = []
    # split text into sentences
    text_sent = nltk.sent_tokenize(text)
    print(len(text_sent), "sentence(s) detected...")
    # for each sentence
    word_reviewed = 0
    for sentence in text_sent:        
        n_count = 0 # num of negative words
        p_count = 0 # num of positive words
        # split each sentence into words
        sent_words = nltk.word_tokenize(sentence)
        for word in sent_words: # loop through each words
            if word in positive_words:
                # prevent clutter
                if word_reviewed <= 2: print("- Identified '" + word + "' as positive")
                p_count += 1 # if word is inside positive word list, add p_count
                word_reviewed +=1
            if word in negative_words: 
                if word_reviewed <= 2: print("- Identified '" +  word + "' as negative")
                n_count += 1 # if word is inside negative word list, add p_count
                word_reviewed +=1
            
        if word_reviewed == 2: print('...')
        
        if p_count == n_count: 
            # 0 means neutral
            sentiment_score_list.append(0)
        elif p_count > n_count : 
            # 1 means positive
            sentiment_score_list.append(1)
        else: 
            # -1 means negative
            sentiment_score_list.append(-1)
    # your sentiment_score_list should be a bunch of numbers like [1, 0, -1, 1, 1] — which you can find the average to see the sentiment score
    return sentiment_score_list
        
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use("ggplot")

# male (male.txt) — get all male names
male_name_list = [('male', name) for name in names.words('male.txt')]
male_name_list = random.sample(male_name_list, 200) # randomly filter to 200 names for quick demo
# female (female.txt) — get all female names
female_name_list = [('female', name) for name in names.words('female.txt')]
female_name_list = random.sample(female_name_list, 200) # randomly filter to 200 names for quick demo
male_name_list
```

```python
# name list consist of all male and female names
name_list = male_name_list + female_name_list
# we are only getting last 2 letters because we want to see the relationship of male to female name in terms of their last 2 letters
last_two_letter_name_list = [(gender, name[-2:]) for (gender, name) in name_list]
name_cfd = nltk.ConditionalFreqDist(last_two_letter_name_list)
```

**Plotting A Graph to see the r/s between male and female name based on last 2 letters**
- The good thing here is you can see that there is a distinct difference between male and female names that you can tap on to build a feature set based on the last 2 letters of their name

```python
plt.figure(figsize=(50,10))
# you can somewhat tell some names is very male-ish and others are very female-ish
name_cfd.plot()
```

```python
def name_features(name):
    return {"pair": name[-2:]}

name_features("Katy")
```

**Before shuffling, all male names are at the front because we add the list using male + female**

```python
name_list[:10]
```

```python
# shuffle the name list in place (i.e. this will edit the variable itself)
random.shuffle(name_list)
```

**After shuffling, male and female names are randomized**

```python
name_list[:10]
```

```python
# for each (gender,name) in name_list i am taking their name_Feautres (i.e. last 2 char) and gender
features = [(name_features(name), gender) for (gender,name) in name_list]
features[:10]
```

```python
len(features)/2 # split all data into 2 — this means that there are 30 * 2 = 60 data points
```

**Create a training and testing set**

```python
training_set = features[:200] # training set takes 30
testing_set = features[200:] # testing set takes 30
```

**Create naive bayes classifier**

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

**Checking the accuracy of our test set shows that it is ~60% accurate**

```python
nltk.classify.accuracy(classifier, testing_set)
```

## Term Frequency, Inverse Document Frequency (TF-IDF)

- This is a method for finding important words and documents when you have a collection of documents and you can use this if you want to determine what a document might be about, if you
- Help you understand the r/s between multiple documents


```python
import nltk
import math

dataset = {
    "tfidf_1.txt":open("tfidf_1.txt", encoding="utf8").read(),
    "tfidf_2.txt":open("tfidf_2.txt", encoding="utf8").read(),
    "tfidf_3.txt":open("tfidf_3.txt", encoding="utf8").read(),
    "tfidf_4.txt":open("tfidf_4.txt", encoding="utf8").read(),
    "tfidf_5.txt":open("tfidf_5.txt", encoding="utf8").read(),
    "tfidf_6.txt":open("tfidf_6.txt", encoding="utf8").read(),
    "tfidf_7.txt":open("tfidf_7.txt", encoding="utf8").read(),
    "tfidf_8.txt":open("tfidf_8.txt", encoding="utf8").read(),
    "tfidf_9.txt":open("tfidf_9.txt", encoding="utf8").read(),
    "tfidf_10.txt":open("tfidf_10.txt", encoding="utf8").read()
}

dataset['tfidf_1.txt']
```

```python
# Calculate term frequencies
def tf(dataset, file_name):
    text = dataset[file_name]
    tokens = nltk.word_tokenize(text)
    fd = nltk.FreqDist(tokens)
    return fd

# Calculate inverse document frequency
def idf(dataset, term):
    count = [term in dataset[file_name] for file_name in dataset]
    inv_df = math.log(len(count)/sum(count))
    return inv_df

def tfidf(dataset, file_name, n):
    term_scores = {}
    file_fd = tf(dataset,file_name)
    for term in file_fd:
        if term.isalpha():
            idf_val = idf(dataset,term)
            tf_val = tf(dataset, file_name)[term]
            tfidf = tf_val*idf_val
            term_scores[term] = round(tfidf,2)
    return sorted(term_scores.items(), key=lambda x:-x[1])[:n]
```

```python
for file_name in dataset:
    print("{0}: \n {1} \n".format(file_name, tfidf(dataset,file_name,5)))
```
